# **Question 2: Undersand TF-IDF and Document representation**

(20 points). Starting from the documents (all the reviews, or abstracts, or tweets) collected for assignment two, write a python program:

(2) To rank the documents with respect to query (design a query by yourself, for example, "An Outstanding movie with a haunting performance and best character development") by using **cosine similarity**.

```python
from scipy._lib.doccer import doc_replace
import numpy
# Write your code here
file_tf = open('stsa-test.txt')
file_tf = file_tf.read()
file_tf = file_tf.split(".")
print(file_tf)
print(len(file_tf))

def doc_freqncy(token):
  if token in file_tf:
    return 1
  else:
    return 0

tfidf = {}
Num = len(file_tf)
for i in range(Num):
  tokens = file_tf[i].split()
  counter = list(set(tokens))
  for token in numpy.unique(tokens):
    tf = tokens.count(token) / len(tokens)
    df = doc_freqncy(token)
    import numpy
    idf = numpy.log(Num/(df+1))
    tfidf[token] = tf*idf

print(tfidf)


def finding_score(query, tokens):
  qry_weights = {}
  for key in tfidf:
    if key in tokens:
      if key in qry_weights:
        qry_weights[key] += tfidf[key]
      else:
        qry_weights[key] = tfidf[key]
  
  return qry_weights
  

sample_query = "anything movement"
curr_tokens = sample_query.split()

qwts = finding_score(sample_query, curr_tokens)
print("Printing the weights of the query here....!!!")
print(qwts)


### Applying Cosine Similarity
import numpy
from numpy.linalg import norm
all_words = []
for f in file_tf:
  for wrd in f.split():
    all_words.append(wrd)

def cos_sim(query):
  query_lst = query.split()
  import numpy
  from numpy.linalg import norm
  cosine_val = numpy.dot(query_lst, all_words) / (norm(query_lst)*norm(all_words))
  print("Cosine similarity is :", cosine_val)
  return cosine_val


# Computing for sample queries
sample_queries = ["no movement", "not much"]
cos_sim_vals = {}
for query in sample_queries:
  cos_val = cos_sim(query)
  cos_sim_vals[query] = cos_val

print("Cosine SImilarity based on queries is as follows : ") 
print(cos_sim_vals)
```

```python
# The GitHub link of your final csv file

# Link: 
import unicodedata
import re, string
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords
from vaderSentiment.vaderSentiment import SentimentIntensityAnalyzer
nltk.download('stopwords')
analyzer = SentimentIntensityAnalyzer()

tweet = ["asd kdbvd bv uhsdbv bsdyc"]
for i in tweet:
  try:
      new_str = re.sub(r'http\S+', '', tweet)
      new_str = unicodedata.normalize("NFKD", new_str)
      new_str = re.sub('[\n\t\v\r]',' ',new_str)
      new_str = new_str.translate(str.maketrans('','',string.punctuation)).lower()
      text_tokens = word_tokenize(new_str)
      new_str_without_tokens = [word for word in text_tokens if not word in stopwords.words()]
      new_str = (" ").join(new_str_without_tokens)
      ps = analyzer.polarity_scores(new_str)

      sentiment = 'neutral'
      if ps['compound'] > 0.2:
          sentiment = 'postive'
      elif ps['compound'] < -0.2:
          sentiment = 'negative'
      else:
          sentiment = 'neutral'
      
      sentiment_values = {'sentiment': sentiment, 'sentiment_score': ps['compound']}
      print(sentiment_values)

  except Exception as e:
      print(f"Error: {e}")
      sentiment_values = {'sentiment': 'Info N/A', 'sentiment_score': 'Info N/A'}



```
